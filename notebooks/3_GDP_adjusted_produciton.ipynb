```python
from owid import catalog
# opening TheWorldBanks's catalogue
plastic_catalogs = catalog.find('plastic')
dataset_name = "plastic_use_polymer"
dataset_path = plastic_catalogs[plastic_catalogs['table'] == dataset_name].path.iloc[0]

# loading method from the docs
rc = catalog.RemoteCatalog() 
plastic_polymer_df = rc[dataset_path]

# adjusting total_plastic_prod_x
total_plastic_prod_x = plastic_polymer_df.loc['World', 2010]['total'] * x_to_world
total_plastic_prod_x
```

```python
yearly_plastic_production_df = pd.DataFrame(columns=['country', 'code', 'year', 'total_plastic_prod'])

years_columns = list(set(gdp_data_of_interest.columns).difference({'Country Name', 'Country Code',
                                              'Indicator Name', 'Indicator Code',
                                              'Unnamed: 68'}))
years_columns.sort()
years_columns = years_columns[30:]

for country in common_countries:
    gdp_country = gdp_data_of_interest[gdp_data_of_interest['Country Name'] == country]
    base_value = gdp_country['2010']
    
    code = waste_generation.loc[waste_generation['country'] == country, 'code'].values[0]
    waste_2010 = waste_generation.loc[waste_generation['country'] == country, 'total_plastic_waste'].values[0]
    
    for year in years_columns:
        proprtion_to_2010 = (gdp_country[year]/base_value).values[0]
        plastic_production = waste_2010 * proprtion_to_2010 / waste_to_production
        data = [[country, code, year, plastic_production]]
        yearly_plastic_production_df = pd.concat([yearly_plastic_production_df, 
                                                pd.DataFrame(data,columns=['country', 'code', 'year', 'total_plastic_prod'])])
```
